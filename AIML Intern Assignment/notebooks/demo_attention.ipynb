```python
import sys
import os

# Update the path to point to the src folder
src_path = os.path.abspath("ml-intern-assessment/AIML Intern assignment/src")
sys.path.append(src_path)

# Now imports work
from attention import scaled_dot_product_attention
from utils import create_padding_mask
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
batch_size, seq_len, d_k, d_v = 1, 6, 8, 6
Q = np.random.rand(batch_size, seq_len, d_k)
K = np.random.rand(batch_size, seq_len, d_k)
V = np.random.rand(batch_size, seq_len, d_v)

# Optional padding mask
mask = create_padding_mask(np.array([[1,1,1,1,0,0]]))

# Compute attention
output, attn_weights = scaled_dot_product_attention(Q, K, V, mask=mask, causal=True, return_weights=True)

# Print results
print("Attention Output:\n", output)
print("\nAttention Weights:\n", attn_weights)

# Visualize attention weights
plt.figure(figsize=(6,5))
plt.imshow(attn_weights[0], cmap='viridis')
plt.colorbar(label='Attention Weight')
plt.title("Scaled Dot-Product Attention Heatmap")
plt.xlabel("Keys")
plt.ylabel("Queries")
plt.show()
```
